# Resize Image

```python
import os
import tensorflow as tf

original_dir = "/content/drive/MyDrive/Dataset/Dataset_Clean/Dataset_Clean_New"
resize_dir = "/content/drive/MyDrive/Dataset/Dataset_Preprocessing/part 2/Dataset_Resize"
target_size = (1920, 1080)

if not os.path.exists(resize_dir):
    os.makedirs(resize_dir)

# Get a list of all image files in the original directory
for subdir, dirs, files in os.walk(original_dir):
    for filename in files:
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            # Construct the input and output file paths
            original_path = os.path.join(subdir, filename)

            # Read the image using TensorFlow
            image = tf.io.read_file(original_path)
            image = tf.image.decode_image(image, channels=3)

            # Resize with padding to 1920x1080
            resized_image = tf.image.resize_with_pad(image, target_size[0], target_size[1])

            # Cast the resized image to uint8
            resized_image = tf.cast(resized_image, tf.uint8)

            # Get the subfolder name
            subfolder_name = os.path.basename(subdir)

            # Create a subfolder under resize_dir for processed images
            output_folder = os.path.join(resize_dir, subfolder_name)
            os.makedirs(output_folder, exist_ok=True)

            # Save the image with padded resizing
            resize_filename = f"resized_{filename}"
            save_path = os.path.join(output_folder, resize_filename)

            # Save the resized image
            tf.io.write_file(save_path, tf.io.encode_jpeg(resized_image).numpy())

            print(f"Resized and saved {filename} to {target_size} with padding in {output_folder}")
```

# Increase Saturation

```python
import os
import tensorflow as tf

resize_dir = "/content/drive/MyDrive/Dataset/Dataset_Preprocessing/part 2/Dataset_Resize"
saturation_dir = "/content/drive/MyDrive/Dataset/Dataset_Preprocessing/part 2/Dataset_Saturation"

if not os.path.exists(saturation_dir):
  os.makedirs(saturation_dir)

# Get a list of all image files in the original directory
  for subdir, dirs, files in os.walk(resize_dir):
    for filename in files:
      if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        # Construct the input and output file paths
        resize_path = os.path.join(subdir, filename)
        saturation_path = os.path.join(saturation_dir, filename)

        # Read the image using TensorFlow
        image = tf.io.read_file(resize_path)
        image = tf.image.decode_image(image, channels=3)

        # Increase image saturation
        saturated_image = tf.image.adjust_saturation(image, saturation_factor=2)

        # Cast the saturized image to uint8
        saturated_image = tf.cast(saturated_image, tf.uint8)

        # Get the subfolder name
        subfolder_name = os.path.basename(subdir)

        # Create a subfolder under saturation_dir for processed images
        output_folder = os.path.join(saturation_dir, subfolder_name)
        os.makedirs(output_folder, exist_ok=True)

        # Save the image
        saturation_filename = f"saturated_{filename[:-4]}.jpg"
        save_path = os.path.join(output_folder, saturation_filename)

        # Save the resized image
        tf.io.write_file(save_path, tf.io.encode_jpeg(saturated_image).numpy())

        print(f"Saturated and saved {filename} in {output_folder}")
```

# Increase Contrast

```python
import os
import tensorflow as tf

saturation_dir = "/content/drive/MyDrive/Dataset/Dataset_Preprocessing/part 2/Dataset_Saturation"
contrast_dir = "/content/drive/MyDrive/Dataset/Dataset_Preprocessing/part 2/Dataset_Contrast"

# Create the output folder if it doesn't exist
if not os.path.exists(contrast_dir):
  os.makedirs(contrast_dir)

# Get a list of all image files in the original directory
  for subdir, dirs, files in os.walk(saturation_dir):
    for filename in files:
      if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        # Construct the input and output file paths
        saturation_path = os.path.join(subdir, filename)
        contrast_path = os.path.join(contrast_dir, filename)

        # Read the image using TensorFlow
        image = tf.io.read_file(saturation_path)
        image = tf.image.decode_image(image, channels=3)

        # Increase all images's contrast
        contrast_image = tf.image.adjust_contrast(image, contrast_factor=2)

        # Cast the contrast image to uint8
        contrast_image = tf.cast(contrast_image, tf.uint8)

        # Get the subfolder name
        subfolder_name = os.path.basename(subdir)

        # Create a subfolder under contrast_dir for processed images
        output_folder = os.path.join(contrast_dir, subfolder_name)
        os.makedirs(output_folder, exist_ok=True)

        # Save the image
        contrast_filename = f"contrast_{filename[:-4]}.jpg"
        save_path = os.path.join(output_folder, contrast_filename)

        # Save the resized image
        tf.io.write_file(save_path, tf.io.encode_jpeg(contrast_image).numpy())

        print(f"Increased {filename} contrast and saved in {output_folder}")
```

```python
import os
import tensorflow as tf

def resize_images(image, target_size=(1920, 1080)):
    image = tf.io.read_file(image)
    image = tf.image.decode_image(image, channels=3)
    resized_image = tf.image.resize_with_pad(image, target_size[0], target_size[1])
    resized_image = tf.cast(resized_image, tf.uint8)
    return resized_image

def adjust_saturation(image):
  saturated_image = tf.image.adjust_saturation(image, saturation_factor=2)
  saturated_image = tf.cast(saturated_image, tf.uint8)
  return saturated_image

def adjust_contrast(image):
  contrast_image = tf.image.adjust_contrast(image, contrast_factor=2)
  contrast_image = tf.cast(contrast_image, tf.uint8)
  return contrast_image
```

```python
def preprocess_image(image_path):
    resize = resize_images(image_path)  # 1920 x 1080
    saturation = adjust_saturation(resize)  # 2
    contrast = adjust_contrast(saturation)  # 2
    return contrast

if __name__ == '__main__':
    # Path to the input image
    image_path = "/content/drive/My Drive/Dataset/Dataset_Clean/Plantacid-Forte/IMG_20231122_161407.jpg"
    image = keras_ocr.tools.read(image_path)

    # Preprocess the input image
    input_image = preprocess_image(image_path)

    # Perform OCR on the image
    predictions = pipeline.recognize([image])

    # Collect recognized text into a list of dictionaries
    results = []
    for text, box in predictions[0]:
        result_entry = {'text': text, 'left': box[0][0], 'top': box[0][1], 'width': box[1][0] - box[0][0], 'height': box[2][1] - box[0][1]}
        results.append(result_entry)

    # Draw bounding boxes on the image
    for entry in results:
        left, top, width, height = entry['left'], entry['top'], entry['width'], entry['height']

        # Convert coordinates to integers
        left, top, width, height = int(left), int(top), int(width), int(height)

        cv2.rectangle(image, (left, top), (left + width, top + height), (0, 255, 0), 2)

# Display the results
plt.imshow(image)
plt.show()
```
